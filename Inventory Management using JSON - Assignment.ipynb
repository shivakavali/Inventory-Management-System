```python
#importing libraries
import time
import json

#Initialising the variables
sale =''

#Loading the Inventory records from JSON to String 
fd = open('Records.json', 'r')
js = fd.read()
fd.close()


#Converting String records to JSON[Dictionaries]
record = json.loads(js)

#Displaying the menu
print("--------------------MENU------------------\n")

for key in record.keys():
    print(key,":" ,record[key]['Name'],"\t|",record[key]['Price'],"\t|",record[key]['Qn'])

print("--------------------------------------------\n")


#Taking user inputs about the product and purchase details
name  = str(input("Enter Your name       : "))
mail  = str(input("Enter Your Email ID   : "))
phone = str(input("Enter Your Phone No  : "))
ui_id = str(input("Enter Product ID     : "))
ui_qn = int(input("Enter Quantity       : "))

#If we are having enough quantity than user want
if (record[ui_id]['Qn']>=ui_qn):
    
    bill = ui_qn*record[ui_id]['Price']
    if bill >= 500 and bill < 1000:
        dbill = int(((ui_qn*record[ui_id]['Price'])*5)/100)
        total =  str(bill-dbill)
        print("--------------------------------")
        print("               Bill             \n")
        print("Name              :",record[ui_id]['Name'])
        print("Price   (Rs)      :",record[ui_id]['Price'], "Rs")
        print("Quantity          :",ui_qn)

        print("-------------------------------")
        print("Billing Amount    : ",record[ui_id]['Price']*ui_qn ,"Rs")
        print("-------------------------------")
        print("Congratulations! You've reached shopping Amount more than 500")
        print("You've won 5% discount\n")
        print("-------------------------------")
        print("Now your Billing Amount is ", total)
        print("-------------------------------") 
    elif (bill >=1000):
        dbill = int(((ui_qn*record[ui_id]['Price'])*10)/100)
        total =  str(bill-dbill)
        print("--------------------------------")
        print("               Bill             \n")
        print("Name              :",record[ui_id]['Name'])
        print("Price   (Rs)      :",record[ui_id]['Price'], "Rs")
        print("Quantity          :",ui_qn)

        print("-------------------------------")
        print("Billing Amount    : ",record[ui_id]['Price']*ui_qn ,"Rs")
        print("-------------------------------")
        print("Congratulations! You've reached shopping Amount more than 1000")
        print("You've won 10% discount\n")
        print("-------------------------------")
        print("Now your Billing Amount is ", total)
        print("-------------------------------")
        
        
    else:
        #Displaying the purchase details
        print("--------------------------------")
        print("               Bill             \n")
        print("Name              :",record[ui_id]['Name'])
        print("Price   (Rs)      :",record[ui_id]['Price'], "Rs")
        print("Quantity          :",ui_qn)

        print("-------------------------------")
        print("Billing Amount    : ",record[ui_id]['Price']*ui_qn ,"Rs")
        print("-------------------------------")

    
    #Updating my records 
    record[ui_id]['Qn'] = record[ui_id]['Qn'] - ui_qn
    
    #generating the sales structure in 
    sale = name+","+mail+","+phone+","+record[ui_id]['Name']+","+str(ui_qn)+","+str(ui_qn*record[ui_id]['Price'])+","+time.ctime()

elif (record[ui_id]['Qn']==0):
    print("Sorry, given Item was not available")
#If we are not having enough quantity
else:
    print("Sorry, We're not having enough Quantity in our Inventory.")
    print("We're having only ", record[ui_id]['Qn'], "Quantity")
    print("--------------------------------")
    
    ch = str(input("Press Y/y to purchase : "))

    if(ch=="Y" or ch=='y'):
        print("--------------------------------")
        print("               Bill             \n")
        print("Name              :",record[ui_id]['Name'])
        print("Price   (Rs)      :",record[ui_id]['Price'], "Rs")
        print("Quantity          :",record[ui_id]['Qn'])

        print("-------------------------------")
        print("Billing Amount    : ",record[ui_id]['Price']*record[ui_id]['Qn'] ,"Rs")
        print("-------------------------------")
        sale = name+","+mail+","+phone+","+record[ui_id]['Name']+","+str(record[ui_id]['Qn'])+","+str(record[ui_id]['Qn']*record[ui_id]['Price'])+","+time.ctime()
        record[ui_id]['Qn'] = 0
    else:
        print("Thanks")

js = json.dumps(record)

#Saving the customer and sales details in a text file
fd = open("JSONsales.txt", 'a')
fd.write(sale)
fd.close()

#Updating the details in the txt file
fd = open("Records.json", 'w')
fd.write(js)
fd.close()

print("-----------------------------------")
print("Thanks for Order, Updated Inventory.")
print("-----------------------------------")
```
